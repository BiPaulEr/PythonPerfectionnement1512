```python
class Vector(object):
     def __init__(self, x, y):
         self.x = x
         self.y = y
     def __repr__(self):
         return f"<Vector: x={self.x}, y={self.y}>"
     def __eq__(self, other):
         return self.x == other.x and self.y == other.y

vector1 = Vector(3, 4)
print(vector1)
vector2 = Vector(3, 4)
print(vector1 == vector2)
```

```python
class Vector3d:
     def __init__(self, x, y, z):
         self.x = x
         self.y = y
         self.z = z
     def __repr__(self):
         return f"<Vector: x={self.x}, y={self.y}, z={self.z}>"
     def __eq__(self, other):
         return self.x == other.x and self.y == other.y and self.z == other.z

vector1 = Vector3d(3, 4, 5)
print(vector1)
vector2 = Vector3d(3, 4, 5)
print(vector1 == vector2)
```

```python
from dataclasses import dataclass

@dataclass
class Vector3D_dataclass:
    x : int
    y : int
    z : int
    
vector1 = Vector3D_dataclass(3, 4, 5)
print(vector1)
vector2 = Vector3D_dataclass(3, 4, 5)
print(vector1 == vector2)
```

```python
from dataclasses import dataclass

@dataclass(frozen=True)
class MyClass:
    x : int
    y : int
    
myclass1 = MyClass(3, 4)
print(myclass1)
myclass2 = MyClass(3, 4)
print(myclass2 == myclass1)

myclass2.x = 6
```

```python
from dataclasses import dataclass

@dataclass(order = True)
class MyClass:
    x : int
    y : int

        
myclass1 = MyClass(3, 4)
print(myclass1)
myclass2 = MyClass(5, 4)
print(myclass2 == myclass1)


print( myclass2 <= myclass1)
print( myclass2 < myclass1)

print( myclass2 > myclass1)
print( myclass2 >= myclass1)
```

```python
from dataclasses import dataclass, field

@dataclass
class MyClass:
    x : int 
    y : list = field(default_factory=list)
    

myclass = MyClass(5)
myclass.y.append("OK")
myclass1 = MyClass(5)
myclass1.y.append("OK")
print(myclass.y, myclass1.y)
```

```python
from dataclasses import dataclass, field

@dataclass
class MyClass:
    x : int 
    checksum : int = field(init=False, compare=False, repr=False)

    def __post_init__(self):
        self.checksum = self.x.__hash__()
    

myclass = MyClass(5)
print(myclass)
```

```python
from dataclasses import dataclass, field

@dataclass(order = True)
class Produit:
    nom : str = field(compare = False)
    prix : float = field(repr = False)
    categorie : str 
    quantite_en_stock : int = field(init = False)
    def __post_init__(self):
        self.quantite_en_stock = 0



produit1 = Produit(nom="Produit A", prix=20.0, categorie="Électronique")
produit2 = Produit(nom="Produit B", prix=10.0, categorie="Maison")
produit3 = Produit(nom="Produit C", prix=20.0, categorie="Électronique")

print(produit1 == produit2)
print(produit1 == produit3)
print(produit1 > produit3)

print(produit1)
```
