```python
class Metaclass(type):
    @classmethod
    def __prepare__(mcs, name, bases, **kwargs):
        namespace = super().__prepare__(name, bases, **kwargs)
        print("__prepare__", mcs, name, bases, namespace)
        return namespace


    def __new__(mcs, name, bases, namespace):
        classe = super().__new__(mcs, name, bases, namespace)
        print("__new__", mcs, name, bases, classe)
        return classe
   
 
    def __init__(cls, name, bases, namespace, **kwargs):
        print("__init__", cls, name, bases, namespace)
        super().__init__(name, bases, namespace)
   
    def __call__(cls, *args, **kwargs):
        print("__call__", cls)
        instance = super().__call__(*args, **kwargs)
        print("__call__", instance)
        return instance
```

```python
class MyClass(metaclass=Metaclass):
    def __new__(cls, *args, **kwargs):
        variable = super().__new__(cls)
        print("__new__", cls, variable)
        return variable

    def __init__(self):
        print('__init__()', self)
```

```python
myclass = MyClass()
```

```python
class Singleton(type):
    liste_singleton = {}
   
    def __call__(cls, *args, **kwargs):
        if cls.__name__ in cls.liste_singleton:
            return cls.liste_singleton[cls.__name__ ]
        cls.liste_singleton[cls.__name__ ] = super().__call__(*args, **kwargs)
        return cls.liste_singleton[cls.__name__ ]


class Sigleton1(metaclass = Singleton):
    pass

singleton1 = Sigleton1()
singleton2 = Sigleton1()
print(singleton2 == singleton1)

class Sigleton2(metaclass = Singleton):
    pass

singleton1 = Sigleton2()
singleton2 = Sigleton2()
print(singleton2 == singleton1)

print(Singleton.liste_singleton)
```
