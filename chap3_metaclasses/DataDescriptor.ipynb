```python
class DataDescriptors:
    def __init__(self, value, name):
        self.name = name
        self.value = value
        
    def __get__(self, obj, objtype):
        print(f"Retrieving value of {obj} for {objtype}")
        return self.value

    def __set__(self, obj, value):
         print(f"Setting value of {obj} with {value}")
         self.value = value
        
class MyClass:
    y = DataDescriptors(5, "y")

myclass = MyClass()
print(myclass)
myclass.y = 10
print(myclass.y)
```

```python
print(myclass.__dict__)
```

```python
myclass2 = MyClass()
print(myclass2.y)
```

```python
class ReadOnly:
    def __init__(self, value):
        self.value = value
        
    def __get__(self, obj, objtype):
        print(f"Retrieving value of {obj} for {objtype}")
        return self.value

    def __set__(self, obj, value):
         pass
        
class MyClass:
    y = ReadOnly(5)

myclass = MyClass()
print(myclass)
myclass.y = 10
print(myclass.y)
```

```python
class DynamiqueAttribute:
    def __init__(self, function, name):
        self.name = name
        self.function = function
        
    def __get__(self, obj, objtype):
        radius = getattr(obj, self.name)
        return self.function(radius)

    def __set__(self, obj, value):
         self.value = value
        
class MyClass:
    radius = 5
    area = DynamiqueAttribute(lambda x: x**2 * 3.14, "radius")


myclass = MyClass()
print(myclass.area)
```

```python
class LazyAttribute:
    def __init__(self, function):
        self.function = function
        self.value = None
        
    def __get__(self, obj, objtype):
        if self.value is None:
            self.value = self.function()
        return self.value

    def __set__(self, obj, value):
         pass
        
class MyClass:
    lazy = LazyAttribute(lambda : 12)


myclass = MyClass()
```
