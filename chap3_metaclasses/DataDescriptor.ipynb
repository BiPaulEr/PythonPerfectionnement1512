```python
class DataDescriptors:
    def __init__(self, value, name):
        self.name = name
        self.value = value
        
    def __get__(self, obj, objtype):
        print(f"Retrieving value of {obj} for {objtype}")
        return self.value

    def __set__(self, obj, value):
         print(f"Setting value of {obj} with {value}")
         self.value = value
        
class MyClass:
    y = DataDescriptors(5, "y")

myclass = MyClass()
print(myclass)
myclass.y = 10
print(myclass.y)
```

```python
print(myclass.__dict__)
```

```python
myclass2 = MyClass()
print(myclass2.y)
```

```python
class ReadOnly:
    def __init__(self, value):
        self.value = value
        
    def __get__(self, obj, objtype):
        print(f"Retrieving value of {obj} for {objtype}")
        return self.value

    def __set__(self, obj, value):
         pass
        
class MyClass:
    y = ReadOnly(5)

myclass = MyClass()
print(myclass)
myclass.y = 10
print(myclass.y)
```

```python
class DynamiqueAttribute:
    def __init__(self, function, name):
        self.name = name
        self.function = function
        
    def __get__(self, obj, objtype):
        radius = getattr(obj, self.name)
        return self.function(radius)

    def __set__(self, obj, value):
         self.value = value
        
class MyClass:
    radius = 5
    area = DynamiqueAttribute(lambda x: x**2 * 3.14, "radius")


myclass = MyClass()
print(myclass.area)
```

```python
class LazyAttribute:
    def __init__(self, function):
        self.function = function
        self.value = None
        
    def __get__(self, obj, objtype):
        if self.value is None:
            self.value = self.function()
        return self.value

    def __set__(self, obj, value):
         pass
        
class Circle:
    lazy = LazyAttribute(lambda : 12)


myclass = Circle()
```

```python
class Pyramide:
    def __init__(self,longeur_base,hauteur):
        self.longeur_base = longeur_base
        self.hauteur = hauteur

    @property
    def aire_base(self):
        """ bientot heure manger"""
        return self.longeur_base ** 2
    @property
    def volume(self):
        return self.aire_base * self.hauteur * (1/3)
    @property
    def perimeter(self):
        return 4*self.longeur_base + 4 * math.sqrt( (math.sqrt(2) * self.aire_base / 2) + self.hauteur **2 ) 

    @property
    def hauteur(self):
        return self.__hauteur
    @hauteur.setter
    def hauteur(self, value):
        if value < 0:
            raise ValueError("")
        self.__hauteur = value

    @property
    def longeur_base(self):
        return self.__longeur_base
    @longeur_base.setter
    def longeur_base(self, value):
        if value < 0:
            raise ValueError("")
        self.__longeur_base = value

pyramide = Pyramide(3, 3)
print(pyramide.perimeter)
print(pyramide.hauteur)
print(pyramide.volume)
print(pyramide.aire_base)
```
