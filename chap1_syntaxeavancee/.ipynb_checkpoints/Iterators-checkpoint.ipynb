```python
class CompteurIterable:
    def __init__(self, begin):
        self.begin = begin
        self.current = begin
    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        self.current -= 1
        return self.current

    def __iter__(self):
        return self

count = CompteurIterable(10)
for i in count:
    print(i)
print(count.current)
for i in count:
    print(i)
```

```python
class CompteurIteratorIterable:
    def __init__(self, begin):
        self.begin = begin
        self.current = begin
    def __next__(self):
        if self.current <= 0:
            raise StopIteration
        self.current -= 1
        return self.current
        
class CompteurIterator:
    def __init__(self, begin):
        self.begin = begin
    def __iter__(self):
        return CompteurIteratorIterable(self.begin)

count = CompteurIterator(5)
for i in count:
    print(i)

for i in count:
    print(i)
```
