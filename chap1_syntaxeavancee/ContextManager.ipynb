```python
class ContextManager:
    def __enter__(self):
        print("Preparation env")
        return "OBJECT"
        
    def __exit__(self, exc_type, exc_instance, traceback):
        print(exc_type, exc_instance)
        print("nettoyage environnement")
        return True

with ContextManager() as test:
    print(test)
    print("Dans le contexte")
    raise ValueError("VALEUR")

print("Hors du contexte")
```

```python
from contextlib import contextmanager

@contextmanager
def mon_contexte():
    try:
        print("__enter__")
        yield "RESSOURCE"  
    finally:
        print("__exit__")


with mon_contexte() as ressource:
    print(f"Utilisation de la {ressource}")
```

```python
class Transaction:
    def begin(self):
        print("Transaction débutée")

    def commit(self):
        print("Transaction validée (commit)")

    def rollback(self):
        print("Transaction annulée (rollback)")


class TransactionManager():
    def __enter__(self):
        self.transaction = Transaction()
        self.transaction.begin()
        
    def __exit__(self, exc_type, exc_instance, tracaback):
        if exc_type:
            self.transaction.rollback()
        else:
            self.transaction.commit()
        return True

with TransactionManager() as t:
    print("ok")
    raise ValueError("")
        
```

```python
import requests 
session = requests.Session()
headers = {"Accept": "text/plain"}
answer = session.get("https://icanhazdadjoke.com/", headers= headers)
if (answer.status_code == 200):
    print(answer.text)
session.close()
```

```python
class SessionManager():
    def __enter__(self):
        self.session = requests.Session()
        return self.session
    def __exit__(self, exc_type, exc_instance, traceback):
        self.session.close()

with SessionManager() as session:
    headers = {"Accept": "text/plain"}
    answer = session.get("https://icanhazdadjoke.com/", headers= headers)
    if (answer.status_code == 200):
        print(answer.text)
```
